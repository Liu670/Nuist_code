```python
import requests
import re


headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36'}


url = 'https://www.baidu.com/s?rtt=4&tn=news&word=阿里巴巴'
res = requests.get(url, headers=headers).text


# 提取网址和标题
p_href = '<h3 class="c-title t t tts-title">.*?<a.*?href="(.*?)"'
href = re.findall(p_href, res, re.S)
p_title = '<h3 class="c-title t t tts-title">.*?>(.*?)</a>'
title = re.findall(p_title, res, re.S)
# 提取日期和来源
p_date = '<span class="c-color-gray2 c-font-normal">(.*?)</span>'
date = re.findall(p_date, res)
p_source = '<span class="c-color-gray c-font-normal c-gap-right">(.*?)</span>'
source = re.findall(p_source, res)


for i in range(len(title)):
    title[i] = re.sub('<.*?>', '', title[i])
    print(str(i + 1) + '.' + title[i] + '(' + source[i] + ' ' + date[i] + ')')
    print(href[i])
```

```python
p_href = '<a.*?href="(.*?)".*?>.*?</a>'
href = re.findall(p_href, res, re.S)
href
```
