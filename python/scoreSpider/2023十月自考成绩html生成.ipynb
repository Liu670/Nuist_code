```python
import pandas as pd
from bs4 import BeautifulSoup
 
path = './exam/10月成绩表.csv'
# 读取文件并转换为数据框
df = pd.read_csv(path,index_col=False)
df = df.drop(df.columns[0], axis=1)

# df.to_html("1.html",encoding="utf-8")
df
```

```python
def write_to_html_file(df, title='', filename='out.html'):
    '''
    Write an entire dataframe to an HTML file with nice formatting.
    '''

    result = '''
<html>
<head>
<style>

    h1 {
        text-align: center;
        font-family: Helvetica, Arial, sans-serif;
    }
    table {
        margin-left: auto;
        margin-right: auto;
    }
    table, th, td {
        border: 1px solid black;
        border-collapse: collapse;
    }
    th, td {
        padding: 5px;
        text-align: center;
        font-family: Helvetica, Arial, sans-serif;
        font-size: 90%;
    }
    table tbody tr:hover {
        background-color: #dddddd;
    }
    .wide {
        width: 90%;
    }

</style>
</head>
<body>
    '''
    div = '''
        <div class="box" style="position: relative;left: 75%;bottom: 15;">
            <label> 
              <input id="txtsearch" type="text">
              <button id="select" onclick="onSearch(document.getElementById('txtsearch').value.trim(),2)">查询</button> 
              <a href="../"><input type="button" value="返回"></a>
            </label>
        </div>
    '''
    result += '<h1> %s </h1> %s ' % (title,div)
    result += df.to_html(classes='wide', escape=False)
    result += '''
</body>

<script type="text/javascript">  
  /**
  * js 前台搜索
  * @param searchContent 搜索的内容
  * @col 要搜索的哪一列，这里是第一列，从0开始数起  
  */

 function onSearch(searchContent,col){
     setTimeout(function(){//因为是即时查询，需要用setTimeout进行延迟，让值写入到input内，再读取  
      thead_html = document.getElementsByTagName("thead")[0].outerHTML
         var storeId = document.getElementsByTagName('table')[0];//获取table的id标识  //获取table的id标识  
         var rowsLength = storeId.rows.length;//表格总共有多少行  
          // 输入框输入的内容
          var v = document.getElementById("txtsearch").value.trim()
          var select = document.getElementById("select");
          select.value = v;
         var searchCol = col;//要搜索的哪一列，这里是第一列，从0开始数起  
         var flag = false; // 是否存在  true
         for(var i=1;i<rowsLength;i++){//按表的行数进行循环，本例第一行是标题，所以i=1，从第二行开始筛选（从0数起）  
            var searchText = storeId.rows[i].cells[searchCol].innerHTML.trim();//取得table行，列的值  
            if(searchText.match(searchContent) || searchText.toUpperCase().match(searchContent.toUpperCase())){//用match函数进行筛选，如果input的值，即变量 key的值为空，返回的是ture，  
              flag = true;
              storeId.rows[i].style.display='';//显示行操作，  
             }else{  
                storeId.style.border="none";
                storeId.rows[i].style.display='none';//隐藏行操作  
             }  
         }
         if(!flag){
          storeId.rows[1].innerHTML="<h2 style='text-align:center;color:red'>查询的结果为空</h2>"
         }
        //  if(v==''){
        //       storeId.innerHTML="输入的内容不能为空,请重新查询！";
        //       storeId.style.border="none";
        //       storeId.style.textAlign="center";
        //       storeId.style.color="RED";
        //     }  
        // }
     },20);//20为延时时间  
 }   
</script> 
</html>
'''
    with open(filename, 'w',encoding="utf-8") as f:
        f.write(result)
```

```python
write_to_html_file(df, title='2023年10月自考成绩一览表', filename='out.html')
```
