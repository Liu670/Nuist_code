```python
import pandas as pd
import requests,time
from bs4 import BeautifulSoup

# 读取xlsx
def read_file():
    path = "zkz.xlsx"
    df = pd.read_excel(path,dtype=str)
    # 获取需要的 姓名、准考证  信息数据
    students = df[["姓名","准考证号"]]
    return students
    

def getDatas(id_,name):
    print("准考证号:{},姓名:{}".format(id_,name))
    url = "https://sdata.jseea.cn/tpl_front/score/allScoreList.html/?zkzh={}&ksmx={}".format(id_,name)
    print(url)
    headers = {
        'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36'
    }
    resp = requests.get(url,headers=headers)
    html = resp.content.decode('utf-8')
    soup = BeautifulSoup(html, "html.parser") 
    table = soup.find_all(name='td')
    j = 1
    datas = []
    scores = {}
    eng=[]
    scores["准考证号"]=id_
    scores["姓名"]=name
    for i in table:
        if i.string != "理论课程":
            datas.append(i.string)
    for i in range(2,len(datas)):
        if len(datas[i])<=3:
            if (eval(datas[i])<=100 and eval(datas[i])>=0) or datas[i]=="免考":
                scores[datas[i-1]] = datas[i]
                if "英语" in datas[i-1]:
                    eng.append(datas[i])
                scores["英语（二）"] = ",".join(eng)

    return scores;

    
# 将数据进行简单的预处理
def process_data(students):
    scores_list = []
    for i in range(0,len(students)):
        scores_dic = getDatas(students["准考证号"][1],students["姓名"][0])
        scores_list.append(scores_dic)
    return scores_list
        
 # 将数据保存为csv文件 
def write_file(scores_list):
    df = pd.DataFrame(scores_list)
    df.to_csv("2023年10月自考成绩.csv",index=False)
    print("文件保存成功!")
    

if __name__=="__main__":
    # 读取文件
    students = read_file()
    # 爬取数据 并处理
    scores_list = process_data(students)
    print(scores_list)
```
