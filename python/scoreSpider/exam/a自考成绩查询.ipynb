```python
import requests
from bs4 import BeautifulSoup

def getDatas(id_,name):
    print("准考证号:{},姓名:{}".format(id_,name))
    url = "https://sdata.jseea.cn/tpl_front/score/allScoreList.html/?zkzh={}&ksmx={}".format(id_,name)
    print(url)
    headers = {
        'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36'
    }
    resp = requests.get(url,headers=headers)
    html = resp.content.decode('utf-8')
    soup = BeautifulSoup(html, "html.parser") 
    table = soup.find_all(name='td')
    j = 1
    datas = []
    s1 = {}
    eng=[]
    s1["准考证号"]=id_
    s1["姓名"]=name
    for i in table:
        if i.string != "理论课程":
            datas.append(i.string)
    for i in range(2,len(datas)):
        if len(datas[i])<=3:
            if eval(datas[i])<=100 or eval(datas[i])>=0 or datas[i]=="免考":
                s1[datas[i-1]] = datas[i]
                if "英语" in datas[i-1]:
                    eng.append(datas[i])
                s1["英语（二）"] = ",".join(eng)

    return s1;
```
